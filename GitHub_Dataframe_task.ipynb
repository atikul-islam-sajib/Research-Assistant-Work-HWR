```python
import pandas as pd

def convert_to_raw_url(github_url):
    if "github.com" in github_url and "/blob/" in github_url:
        raw_url = github_url.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")
        return raw_url
    else:
        raise ValueError("The provided URL is not a valid GitHub file URL.")

# Original GitHub URL
github_url = "https://github.com/markusloecher/DataScience2018/blob/master/data/Carseats.csv"

# Convert to raw URL
raw_url = convert_to_raw_url(github_url)

# Read the CSV file
cervical = pd.read_csv(raw_url)

# Display the first few rows to confirm it worked
cervical.head()
```
